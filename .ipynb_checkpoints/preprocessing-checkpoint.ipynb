## Cleaning the data

```python
import string
import numpy as np
import pandas as pd
```

```python
def object_values_to_int(x):
    new_val=''
    if type(x) == str:
        for i in x:
            if i in string.digits:
                new_val +=i
        return int(new_val)
    return x
```

```python
#loop through every value #len(df15.columns) == 19 #len(df15.index) == 50
for i in range(0,50):
    for j in range(1,19):
        df15.iloc[i,j] = object_values_to_int(df15.iloc[i,j])
```

```python
#change all numeric dtypes to int
for header in headers:
    if header !='Counties':
        df15[header] = df15[header].astype(int)
```

```python
#remove unwanted rows and columns
df15.drop([47,48,49], inplace=True)
```

```python
df15.drop('Grand total', inplace=True, axis=1)
```

```python
#make the counties column the index
df15=df15.set_index('Counties')
```

```python
#add an year column since we may combine all dataframes
df15['year']='2015'
df15.head()
```

```python
# **mistake: rename one column
df15=df15.rename(columns={'Theft of vehicles and other parts': 'Other offences against persons'})
df15.head()
```

```python
df15.to_csv('2015.csv')
```
